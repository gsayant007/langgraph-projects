```python
import numpy as np
import matplotlib.pyplot as plt

class Vector2:
    def __init__(self,x:float,y:float):
        self.x = x
        self.y = y
class Ball_2(Vector2):
    def __init__(self,position:Vector2,velocity:Vector2):
        self.position = position
        self.velocity = velocity

class Ball:
    def __init__(self,x:float,y:float,vx:float,vy:float):
        self.x = x
        self.y = y
        self.vx = vx
        self.vy = vy

def updateBall(p:Ball_2,dt:float,g:float):
    p.position.x =p.position.x + p.velocity.x * dt
    p.position.y = p.position.y + p.velocity.y * dt
    p.velocity.y = p.velocity.y - g*dt

def plot_trajectory(arr:np.array):
    fig,ax = plt.subplots()
    ax.plot(arr[:,0],arr[:,1])
    ax.grid(visible=True)
    plt.tight_layout()
    plt.show()

def main():
    dt = 0.1
    g = 9.8
    # x = 0.0
    # y = 0.0
    # vx = 10.0
    # vy = 20.0
    pos = Vector2(0.0,0.0)
    vel = Vector2(10.0,20.0)
    global trajectory
    trajectory = []
    # ball = Ball(x,y,vx,vy)
    ball = Ball_2(pos,vel)
    for t in range(0,50):
        print(f"t={t}, x={ball.position.x}, y={ball.position.y:.3f}")
        trajectory.append((ball.position.x,ball.position.y))
        updateBall(p=ball,dt=dt,g=g)
        if(ball.position.y<0):
            break
    trajectory = np.array(trajectory)
    plot_trajectory(trajectory)
    # print(trajectory,end="\n")

if __name__ == "__main__":
    main()
```
